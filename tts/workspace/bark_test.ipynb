```python
from transformers import AutoProcessor, BarkModel

processor = AutoProcessor.from_pretrained("suno/bark")
model = BarkModel.from_pretrained("suno/bark")

# voice_preset = "v2/en_speaker_6"
voice_preset = "v2/ko_speaker_0"

# inputs = processor("안녕하세요?", voice_preset=voice_preset)
# audio_array = model.generate(**inputs)
```

```python
inputs = processor("안녕하세요?", voice_preset=voice_preset)
# Add attention_mask and pad_token_id
# inputs['attention_mask'] = inputs.input_ids.ne(processor.tokenizer.pad_token_id)
# inputs['pad_token_id'] = processor.tokenizer.eos_token_id
```

```python
audio_array = model.generate(**inputs)
```

```python
from IPython.display import Audio

sample_rate = model.generation_config.sample_rate
Audio(audio_array, rate=sample_rate)
```
