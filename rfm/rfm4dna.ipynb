## 일대여비용 분포 확인

```python
from IPython.display import clear_output

# Create an IntSlider widget
cost_limit_slider = widgets.IntSlider(
    value=50000,
    min=0,
    max=100000,
    step=1000,
    description='Cost Limit:',
    continuous_update=False
)
# Create a Button widget
apply_button = widgets.Button(description='Apply')


def on_button_clicked(b):
    # This function will be called when the button is clicked
    # It redraws the chart with the new cost limit
    cost_limit = cost_limit_slider.value
    clear_output(wait=True)  # Clear the output
    display(cost_limit_slider, apply_button)  # Display the slider and button again
    show_ma_distribution(ht=400, cost_limit=cost_limit)

apply_button.on_click(on_button_clicked)
display(cost_limit_slider, apply_button)
```
